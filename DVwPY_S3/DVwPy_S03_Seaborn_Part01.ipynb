## <p style="background-color:#E51A59; font-family:newtimeroman; color:#FFF9ED; font-size:150%; text-align:center; border-radius:10px 10px;">MATPLOTLIB vs SEABORN</p>

<a id="0"></a>
<a href="#toc" class="btn btn-primary btn-sm" role="button" aria-pressed="true" 
style="color:blue; background-color:#dfa8e4" data-toggle="popover">Content</a>

- ### 1.Functionality:

  - **``Matplotlib:``** Matplotlib is mainly deployed for **``basic plotting``**. Visualization using Matplotlib generally consists of bars, pies, lines, scatter plots and so on.

  - **``Seaborn:``** Seaborn, on the other hand, provides a variety of visualization patterns. It uses **``fewer syntax``** and has **``easily interesting default themes``**. It specializes in **``statistics visualization``** and is used if one has to summarize data in visualizations and also show the distribution in the data.

- ### 2.Handling Multiple Figures:

  - **``Matplotlib:``** Matplotlib has **``multiple figures``** that can be opened, but **``need to be closed explicitly``**. **``plt.close()``** only closes the current figure. **``plt.close(‚Äòall‚Äô)``** would close them all.

  - **``Seaborn:``** Seaborn **``automates``** the creation of **``multiple figures``**. This sometimes leads to **``OOM (out of memory) issues``**.

- ### 3.Visualization:

  - **``Matplotlib:``** Matplotlib is a graphics package for data visualization in Python. It is **``well integrated with NumPy and Pandas``**. The **``pyplot module``** mirrors the **``MATLAB``** plotting commands closely. Hence, MATLAB users can easily transit to plotting with Python.

  - **``Seaborn:``** Seaborn is more integrated for working with **``Pandas DataFrames``**. It extends the Matplotlib library for creating beautiful graphics with Python using a **``more straightforward set of methods``*.

- ### 4.Data frames and Arrays

  - **``Matplotlib:``** Matplotlib **``works with DataFrames and arrays``**. It has different stateful APIs for plotting. The **``figures``** and **``axes``** are represented by the **``object``** and therefore plot() like calls without parameters suffices, without having to manage parameters.

  - **``Seaborn:``** Seaborn **``works with the dataset as a whole``** and is **``much more intuitive than Matplotlib``**. For Seaborn, replot() is the entry API with ‚Äòkind‚Äô parameter to specify the type of plot which could be line, bar, or many of the other types. Seaborn is **`not stateful``**. Hence, plot() would require passing the object.

#### <p style="background-color:#E51A59; font-family:newtimeroman; color:#FFF9ED; font-size:150%; text-align:LEFT; border-radius:10px 10px;">How to change a figure's size in Python Seaborn package?</p>

In Seaborn, we can change a figure's size using the **``plt.figure()``** function from Matplotlib or by setting the **``figsize``** parameter within the plotting functions.
- **``For a Single Cell:``**
  - We can set the figure size directly in a specific plotting cell using **``plt.figure(figsize=(width, height))``**.
    - ``plt.figure(figsize=(12, 6))``  # Width = 12, Height = 6 
- **``For the Whole Notebook:``**
  - To set the default figure size for all plots in a notebook, we can use the **``set_context()``** function OR **``set the rcParams``** in Matplotlib. Here are two methods:
    - Method 1: Using set_context()
      - ``sns.set_context("notebook", rc={"figure.figsize": (12, 6)})``  # Default size for all plots
    - Method 2: Using rcParams
      - ``plt.rcParams["figure.figsize"] = (12, 6)``  # Width = 12, Height = 6

The commands **``plt.figure(figsize=(12, 8))``** and **``plt.rcParams['figure.figsize'] = (12, 8)``** in Matplotlib serve different purposes in managing figure sizes. 
- The **``former``** creates a specific figure with the designated dimensions, **``applying only to that figure``** and **``NOT affecting subsequent plots``**.
- In contrast, the **``latter``** sets a **``default figure size for ALL plots``** created **``in the current session``**, ensuring consistency across visualizations until changed. 

Thus, the **``first command``** offers **``localized control``**, while the **``second``** provides a **``global setting``** for figure dimensions.

**The following code sets a ``default figure size for all subsequent plots`` using Seaborn's set() function.**

```python
# rc -- > Run Configuration

# Affect all figures created after this setting in the current session until we change it again or reset it
sns.set(rc={'figure.figsize':(10, 6)})  # specifies the default figure size for all subsequent plots

# Create the scatter plot
sns.scatterplot(x='total_bill', y='tip', data=tips);
```

### <p style="background-color:#E51A59; font-family:newtimeroman; color:#FFF9ED; font-size:150%; text-align:LEFT; border-radius:10px 10px;">"size" vs "s" Parameters</p>

<a id="3.2"></a>
<a href="#toc" class="btn btn-primary btn-sm" role="button" aria-pressed="true" 
style="color:blue; background-color:#dfa8e4" data-toggle="popover">Content</a>

Both **``size``** and **``s``** parameters relate to the size of the markers (points) in the plot, but they **``serve different purposes``**:

- **``size``** parameter allows us to **``size based on another column``**. It groups variable that will **``produce points with different sizes``**. It can be either **``categorical``** or **``numeric``**, although size mapping will behave differently in latter case. Size parameter maps marker sizes to a variable in the dataset, allowing for a **``dynamic representation of data across different points``**. 
- On the other hand **``s``** parameter **``sets a static size for all markers``**, making it easier to control the appearance of the plot without regard to data values. The **``s``** parameter in Seaborn's plotting functions is designed for **``continuous variables``**.

**``size`` Parameter:**
- **``Purpose``**: The **``size``** parameter is used to map the sizes of the markers to a variable in our dataset.
- **``Type``**: Typically expects a **``continuous numerical variable``**.
- **``Functionality``**: When we specify a variable for size, Seaborn will scale the marker sizes based on the values in that variable. This allows for a visual representation of another dimension of data.
- **``Example Use Case``**: We might want to represent the size of the markers based on the total number of people in a group (e.g., size='size').
  - **``Example``**: ``sns.scatterplot(x='total_bill', y='tip', data=tips, size='size', sizes=(20, 200));``

**``s`` Parameter:**
- **``Purpose``**: The **``s``** parameter is used to set a **``fixed size``** for all markers in the scatter plot.
- **``Type``**: Expects a **``numerical value``** (or an array of numerical values for individual sizes).
- **``Functionality``**: It defines the size of the markers uniformly across the plot, meaning **``all points will be the same size``** unless specified otherwise.
- **``Example Use Case``**: We may want all points to be larger for better visibility, regardless of any data variable.
  - **``Example``**: ``sns.scatterplot(x='total_bill', y='tip', data=tips, s=100);``  # All points have the same size of 100

### <p style="background-color:#E51A59; font-family:newtimeroman; color:#FFF9ED; font-size:150%; text-align:left; border-radius:10px 10px;">The Number of Bins in Histogram</p>

The choice of bins for computing and plotting a histogram can exert substantial influence on the insights that one is able to draw from the visualization. If the bins are too large, they may erase important features. On the other hand, bins that are too small may be dominated by random variability, obscuring the shape of the true underlying distribution. The default bin size is determined using a reference rule that depends on the sample size and variance. This works well in many cases, (i.e., with ‚Äúwell-behaved‚Äù data) but it fails in others. It is always a good to try different bin sizes to be sure that you are not missing something important. This function allows you to specify bins in several different ways, such as by setting the total number of bins to use, the width of each bin, or the specific locations where the bins should break.

<p style="background-color:#FFFFFF; font-family:newtimeroman; color:#0000FF; font-size:125%; text-align:left;"><strong>üëâ Modifying bin number in Histogram based on Sturge‚Äôs rule:</strong></p>


**The code below uses Sturge‚Äôs rule to calculate the optimal number of bins for a histogram, which is particularly useful for visualizing the distribution of the "total_bill" variable in the tips dataset. Here‚Äôs a breakdown of the code, along with an example implementation:**

```python
# Method 1: Sturge‚Äôs rule

# Calculate the optimal number of bins
bin_count = int(np.ceil(np.log2(len(tips.total_bill))) + 1)  

# Create a new figure with the specified size
fig = plt.figure(figsize=(7, 7))

# Create the histogram
sns.histplot(x=tips.total_bill, bins=bin_count)

# Set the title
plt.title(f'bins = {bin_count}');
```

- **``Sturge‚Äôs Rule``**: This method helps determine a reasonable number of bins for a histogram based on the size of the dataset, which can improve the clarity of the visualization.



<p style="background-color:#FFFFFF; font-family:newtimeroman; color:#0000FF; font-size:110%; text-align:left;"><strong>-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</strong></p>

<p style="background-color:#FFFFFF; font-family:newtimeroman; color:#0000FF; font-size:125%; text-align:left;"><strong>üëâ Modifying bin number in Histogram based on Freedman-Diaconis rule:</strong></p>

**The following code implements the ``Freedman-Diaconis rule`` to determine the ``optimal number of bins`` for a histogram of the total_bill amounts in the tips dataset.**

```python
# Method 2: Freedman-Diaconis rule

q1 = tips.total_bill.quantile(0.25)  # Calculates the the 25th percentile (Q1)
q3 = tips.total_bill.quantile(0.75)  # Calculates the 75th percentile (Q3)
iqr = q3 - q1  # Calculates the difference between the 75th percentile (Q3) and the 25th percentile (Q1)
bin_width = (2 * iqr) / (len(tips.total_bill) ** (1 / 3))  # The formula for bin width
bin_count = int(np.ceil((tips.total_bill.max() - tips.total_bill.min()) / bin_width))  #  The calculated number of bins

# Create a new figure with the specified size
fig = plt.figure(figsize=(7, 7))

# Create the histogram
sns.histplot(x=tips.total_bill, bins=bin_count)

# Set the title
plt.title(f'bins = {bin_count}');
```

- **``Freedman-Diaconis Rule``**: This method provides a robust way to determine the number of bins by focusing on the IQR, making it effective for datasets with outliers or skewed distributions.

## <p style="background-color:#E51A59; font-family:newtimeroman; color:#FFF9ED; font-size:150%; text-align:center; border-radius:10px 10px;">THE KERNEL DENSITY ESTIMATION (KDE) PLOT</p>

A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analagous to a histogram. KDE represents the data using a continuous probability density curve in one or more dimensions. The approach is explained further in the [user guide](https://seaborn.pydata.org/tutorial/distributions.html#tutorial-kde).

The KDE plot maps an estimate of a probability *density* function of a random variable. Kernel density estimation is a fundamental data smoothing problem where inferences about the population are made, based on a finite data sample.

- üßê **[seaborn.kdeplot](https://seaborn.pydata.org/generated/seaborn.kdeplot.html#:~:text=A%20kernel%20density%20estimate%20(KDE,further%20in%20the%20user%20guide.)**


<p style="background-color:#FFFFFF; font-family:newtimeroman; color:#FF0000; font-size:125%; text-align:left;"><strong>ü§îüí≠ How does a Kernel Density Estimation work‚ùì</strong></p>


- The Kernel Density Estimation works by plotting out the data and beginning to create a curve of the distribution. The **``curve``** is calculated by **``weighing the distance of all the points in each specific location along the distribution``**. If there are **``more points grouped locally``**, the **``estimation is higher as the probability of seeing a point at that location increases``**. The **``kernel function``** is the specific mechanism used to **``weigh the points``** across the data set. The **``bandwidth of the kernel changes its shape``**. A lower bandwidth limits the scope of the function and leads to the estimate curve looking rough and jagged. By tweaking the parameters of the kernel function (bandwidth and amplitude), one changes the size and shape of the estimate.


<p style="background-color:#FFFFFF; font-family:newtimeroman; color:#FF0000; font-size:125%; text-align:left;"><strong>ü§îüí≠ Kernel Density Estimation and Machine Learning:</strong></p>


- The Kernel Density Estimation technique can be incorporated into [machine learning](https://deepai.org/machine-learning-glossary-and-terms/machine-learning) applications. For example, as the estimation function has parameters to define the scope of the kernel, a [neural network](https://deepai.org/machine-learning-glossary-and-terms/neural-network) can begin to train itself to correct its estimations and produce more accurate results. As the estimation process repeats itself, the bandwidth and amplitude estimations are continuously updated to increase the accuracy of the estimated probability density curve.

the **``KDE``** is constructed through the **``summation``** of the **``kernel``** (most commonly [Gaussian](https://en.wikipedia.org/wiki/Normal_distribution)), we can effect the **``bandwith``** of this kernel **to make the KDE more "sensitive" to the data.** Notice how with a **``smaller bandwith``**, the **``kernels do NOT stretch so wide``**, meaning we do NOT need the cut-off anymore. This is **``analagous to increasing the number of bins in a histogram``** (making the actual bins narrower).

In statistics, **``Kernel Density Estimation``** is a **``non-parametric``** technique that calculates and plots the **``probability distribution (the probability density) of a continuous random variable``**. i.e., The calculation does **``NOT assume the underlying data to be following the assumptions of a normal distribution or any distribution``**.
In simple terms, Kernel Density Estimate is like a **``smoothened counterpart of a histogram without the line of histogram intervals and their end-points``**.

<p style="background-color:#FFFFFF; font-family:newtimeroman; color:green; font-size:125%; text-align:left;">üëâ Kernel Density Smoothing, also known as Kernel Density Estimation (KDE), replaces each sample point with a Gaussian-shaped Kernel, then obtains the resulting estimate for the density by adding up these Gaussians.</p>

Such a **``smoothened curve``** for the probability density of a given data is obtained by drawing **``individual estimates for the data points``** and **``summing them up to produce the final contour``**.

<p style="background-color:#FFFFFF; font-family:newtimeroman; color:#0000FF; font-size:120%; text-align:left;"><strong>Key Concepts:</strong></p>


1. **``Kernel Density Estimate (KDE)``**:
   - A KDE is a non-parametric way to estimate the probability density function of a random variable. It smooths the data by placing a kernel (a small, smooth function) at each data point and summing these contributions to create a continuous density curve.
2. **``Bandwidth``**:
   - The bandwidth determines the width of the kernel used in the KDE. It controls how much the data points influence the shape of the density curve.
   - A smaller bandwidth leads to a tighter kernel around each data point, which means that:
     - Each data point's contribution to the density estimate is more localized.
     - The resulting density curve can capture more of the local variations in the data, leading to a curve that closely follows the actual distribution of the data points.
3. **``Wiggly Curve``**:
   - When the bandwidth is small (like in your example with bw_adjust=0.1), the KDE can show more fluctuations and peaks, resulting in a "wiggly" appearance.
   - This wiggly nature indicates that the KDE is sensitive to the variations in the data. While it can provide a more accurate representation of the data distribution, it may also lead to overfitting, where the estimate captures noise rather than the underlying distribution.
4. **``Visual Example``**:
   - Small Bandwidth: The KDE may show several peaks and troughs, reflecting every little fluctuation in the data.
   - Large Bandwidth: The KDE becomes smoother and more generalized, potentially obscuring important features of the distribution, uch as multiple modes (peaks).

```python
from scipy import stats

# Create a figure with specified size and resolution
plt.figure(figsize=(8, 4), dpi=200)

# Calculate bandwidth based on the standard deviation and size of x
bandwidth = x.std() * x.size ** (-0.001)

# Define the support for the kernels
support = np.linspace(-5, 5, 100)

kernels = []

plt.figure(figsize=(12 ,6))

# Loop through each data point in x
for x_i in x:
    kernel = stats.norm(x_i, bandwidth).pdf(support)
    kernels.append(kernel)
    plt.plot(support, kernel, color="lightblue")

# Add a rug plot for the original data
sns.rugplot(x, color="darkblue", linewidth=4);
```

In the context of statistics and data analysis, a kernel refers to a **``function``** used to **``estimate the probability density function (PDF) of a random variable``**. Here‚Äôs a more detailed explanation:

- **``What is a Kernel?``**
  - **``Definition``**: A kernel is a smooth, continuous function used to create a weighted estimate of the likelihood of a data point occurring at a certain value. It helps in smoothing the data points to create a continuous probability distribution.
- **``Kernel Density Estimation (KDE):``**
  - Kernels are often used in Kernel Density Estimation, a non-parametric way to estimate the PDF of a random variable. KDE takes a set of data points and places a kernel (like a Gaussian function) over each point.
The resulting sum of these kernels creates a smooth curve that represents the estimated density of the data.
- **``Common Kernel Functions:``**
  - **``Gaussian Kernel``**: The most commonly used kernel, shaped like a bell curve. It is defined by its mean and standard deviation.
  - **``Epanechnikov Kernel``**: A parabolic-shaped kernel that is optimal for minimizing mean integrated squared error.
  - **``Uniform Kernel``**: A constant kernel that assigns equal weight to all points within a certain range.
- **``Purpose of Using Kernels``**
  - **``Smoothing``**: Kernels help to smooth out the noise in the data, providing a clearer picture of the underlying distribution.
  - **``Flexibility``**: Unlike histogram-based methods, which can be sensitive to bin size, kernels allow for a more flexible and continuous representation of data distributions.
Insight``**: By visualizing the density estimate using kernels, you can gain insights into the shape, central tendency, and spread of the data.

```python
from matplotlib.patches import Patch
from matplotlib.lines import Line2D

# Create a new figure and axis
fig, ax = plt.subplots(figsize=(12, 8))

# Create a KDE plot for 'tip' with hue based on 'size'
ax = sns.kdeplot(data=tips, x="tip", hue="size", fill=True)

# Add vertical lines for the mean and median of tips
ax.axvline(x=tips.tip.mean(), color='r', linestyle=':', lw=3)
ax.axvline(x=np.median(tips.tip), color='b', linestyle=':', lw=3);

# Create custom legend elements
legend_elements = [Line2D([0], [0], marker='s', color='blue', label='Size 1', markersize=15),
                   Line2D([0], [0], marker='s', color='lightblue', label='Size 2', markersize=15),
                   Line2D([0], [0], marker='s', color='red', label='Size 3', markersize=15),
                   Line2D([0], [0], marker='s', color='darkgreen', label='Size 4', markersize=15),
                   Line2D([0], [0], marker='s', color='lightgreen', label='Size 5', markersize=15),
                   Patch(facecolor='orange', edgecolor='r', label='Size 6')]
# Add the first legend for size types
first_legend = plt.legend(handles=legend_elements, title='Size Type', bbox_to_anchor=(0.55, 1))
plt.gca().add_artist(first_legend)

# Add a second legend for mean and median
second_legend = plt.legend(title='Tips Mean & Median', bbox_to_anchor=(0.75, 1), labels=['Tips Mean', 'Tips Median']) 
plt.gca().add_artist(second_legend);
```

### <p style="background-color:#FFFFFF; font-family:newtimeroman; color:#0000FF; font-size:115%; text-align:left;"><strong>üéÅ BONUS USER DEFINED FUNCTION:</strong></p>


<p style="background-color:#FFFFFF; font-family:newtimeroman; color:#0000FF; font-size:125%; text-align:left;"><strong>üëâ Creating a User Defined Functions (DEFs):</strong></p>

**The following function provides a ``histogram`` and a ``box plot`` to help users understand the ``frequency distribution``, ``central tendency``, and ``variability`` of the data before performing further statistical analysis.**

```python
def show_distribution(col):
    
    '''
    
    This function will prints a Histogram and box plot which are graphical representations 
    for the frequency of numeric data values. It aims to describe the data and explore 
    the central tendency and variability before using advanced statistical analysis techniques. 
    
    '''
    # Get statistics
    from termcolor import colored

    print(colored('Statistical Calculations :', 'red', attrs=['bold']))
    print(colored('-'*26, 'red', attrs=['bold']))    
    min_val = col.min()
    max_val = col.max()
    mean_val = col.mean()
    med_val = col.median()
    mod_val = col.mode()[0]

    print(colored('Minimum:{:>7.2f}\nMean:{:>10.2f}\nMedian:{:>8.2f}\nMode:{:>10.2f}\nMaximum:{:>7.2f}\n'.format(min_val,
                                                                                             mean_val,
                                                                                             med_val,
                                                                                             mod_val,
                                                                                             max_val), 'blue', attrs=['bold']))
    
    

    # Create a figure for 2 subplots (2 rows, 1 column)
    fig, ax = plt.subplots(2, 1, figsize=(9, 7))

    # Plot the histogram   
    ax[0].hist(col, bins=30)
    ax[0].set_ylabel('Frequency', fontsize=10)

    # Add lines for the mean, median, and mode
    ax[0].axvline(x=min_val,  color='yellow',     linestyle='dashed', linewidth=2, label='Minimum')
    ax[0].axvline(x=mean_val, color='lightgreen', linestyle='dashed', linewidth=2, label='Mean')
    ax[0].axvline(x=med_val,  color='cyan',       linestyle='dashed', linewidth=2, label='Median')
    ax[0].axvline(x=mod_val,  color='purple',     linestyle='dashed', linewidth=2, label='Mode')
    ax[0].axvline(x=max_val,  color='red',        linestyle='dashed', linewidth=2, label='Maximum')
    ax[0].legend(loc=[0.7, 0.5], fontsize=8)

    # Plot the boxplot   
    medianprops = dict(linestyle='-', linewidth=3, color='m')
    boxprops=dict(linestyle='-', linewidth=1.5)
    meanprops={"marker":"d", "markerfacecolor":"red", "markeredgecolor":"black", "markersize":"10"}
    flierprops={'marker': 'o', 'markersize': 8, 'markerfacecolor': 'fuchsia'}
    
    ax[1].boxplot(col, 
                  vert=False,
                  notch=True, 
                  patch_artist=False,
                  medianprops=medianprops,
                  flierprops=flierprops,
                  showmeans=True,
                  meanprops=meanprops)
    
    ax[1].set_xlabel('value', fontsize=10)
    

    # Add a title to the Figure
    fig.suptitle('Data Distribution', fontsize=15)
```

- **``Purpose``**: The show_distribution function is designed to help users understand the **``distribution of a numeric dataset``** through visualizations and **``statistical summaries``**.
- **``Visual Output``**: It generates a **``histogram``** and a **``box plot``**, allowing for a comprehensive view of the **``data's distribution``**, **``central tendency``**, and **``variability``**.
- **``Statistical Insights``**: The function also provides important **``statistical information``**, which is useful for exploratory data analysis (EDA) prior to more complex analyses.

<p style="background-color:#FFFFFF; font-family:newtimeroman; color:#0000FF; font-size:125%; text-align:left;"><strong>üëâ Creating a User Defined Functions (DEFs):</strong></p>

**The following ``show_density`` function is designed to create and display a ``density plot for a numeric variable`` using a ``kernel density estimate (KDE)``.**

```python
def show_density(col):
    
    '''
    
    This function plots a 'density plot' which is a representation of the distribution of a numeric variableand display it. 
    It uses a kernel density estimate to show the probability density function of the variable.
    
    '''    
    from matplotlib import pyplot as plt

    fig = plt.figure(figsize=(15, 5))

    # Plot density
    col.plot.density()

    # Add titles and labels
    plt.title('Data Density', fontsize=15)

    # Show the mean, median, and mode
    plt.axvline(x=col.mean(),    color='cyan',   linestyle='dashed', linewidth=2, label='Mean')
    plt.axvline(x=col.median(),  color='red',    linestyle='dashed', linewidth=2, label='Median')
    plt.axvline(x=col.mode()[0], color='yellow', linestyle='dashed', linewidth=2, label='Mode')
    plt.legend()

    # Show the figure
    plt.show()
```

- **``Purpose``**: The show_density function **``visualizes the distribution of a numeric dataset``** using a density plot and highlights important statistical measures (mean, median, and mode).
- **``Visual Output``**: The resulting plot provides a **``smooth representation of the data's distribution``**, helping to identify patterns and central tendencies.
- **``Statistical Insights``**: By marking the **``mean``**, **``median``**, and **``mode``**, the function helps users understand the **``central tendency of the data``** in relation to its distribution.
