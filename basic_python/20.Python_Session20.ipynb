## <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:175%; text-align:center; border-radius:10px 10px;">Program Execution</p>
 

<hr>

Python program execution is the process of running a Python program on a computer. It involves a series of steps, from compiling the code into bytecode to executing it on the interpreter. This process is fundamental to understand how Python works and how to write efficient and effective code. 

![python.exe](https://drive.google.com/uc?id=1c07xSRpabcCVCIZUL1dQ-JZvp1Y3Bmm2)

<hr>

The **"bytecode compilation"** process performed by the Python interpreter is different from traditional compilation processes used in compiled languages. In compiled languages, a compiler is used to directly translate **"source code"** (the text-based code that the programmer writes) into machine language (platform-specific processor code). However, the Python interpreter does not directly translate source code into machine language. Instead, it creates an intermediate language called **bytecode.** Therefore, after the Python interpreter performs "bytecode compilation" on the source code, the Python code is transformed into an intermediate bytecode format which is then executed by the Python virtual machine.

As a result, the "bytecode compilation" process performed by the Python interpreter is a faster and less resource-intensive process compared to traditional compilation processes. However, the bytecode generated by this process still needs to be interpreted every time the code is executed, which means that it does not provide the same performance benefits as traditional compiled languages.

In the first step of the execution process that is shown above, the Python interpreter reads and parses the code. While the program is running, the interpreter reads each line of the program one by one, interprets it, and prints the results to the screen or a file. 

The execution of Python programs happens from top to bottom. That is, each line of the program is interpreted and processed sequentially from the beginning to the end. However, this may vary in some special cases. For example, function calls and loop structures can change the flow of the program and affect its execution. 

During this step, the interpreter checks the syntax and structure of the code to ensure that it follows the Python language rules. If there are any errors or syntax mistakes, the interpreter raises an error, produces error messages and halts the program's execution.  Error messages are important for programmers to identify and correct errors.

<hr>

Once the code has been parsed, the interpreter compiles it into bytecode. Bytecode is a lower-level representation of the code that can be executed more efficiently by the interpreter. The bytecode is then stored in memory and used to execute the program.

<hr>

Finally, the interpreter executes the bytecode instruction by instruction, following the program's control flow. This process involves loading and manipulating data, calling functions and methods, and making decisions based on conditions and loops. As the program executes, it can also interact with the environment, accessing files, databases, network connections, and other resources.

This stage where the Python program is executed instruction by instruction, following the program's control flow and interacting with the environment, is performed during the execution of Python programs on the **Virtual Machine (VM)**. 

The Python bytecode, which is the executable machine code, is generated by the Python interpreter and executed by the VM. The VM interprets the bytecode, performs operations, and follows the program's control flow. Additionally, it can interact with the environment, accessing files, databases, network connections, and other resources. Therefore, Python programs involve a collaboration between the interpreter and the VM during their execution.

<hr>

Understanding how Python executes programs is crucial for writing efficient and effective code. It involves parsing, compiling, and executing the code, as well as interacting with the environment. By mastering the execution process, developers can optimize their programs and ensure they perform as intended.

## <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:175%; text-align:center; border-radius:10px 10px;">Error (Sytax errors)</p>
 

<hr>

Python is a high-level programming language that is widely used for its simplicity and readability. However, like any programming language, it has a specific syntax that must be followed for the code to run properly. A syntax error occurs when the code violates one or more of the language's syntax rules.

<hr>

In Python, syntax errors stop the execution of your program and display an error message. This error message indicates which line and character the error occurred on, as well as what type of error it is. Syntax errors are typically caused by a command being misspelled or an expression being formatted incorrectly. Therefore, in order for your program to run properly, you will need to correct any syntax errors.

In Python, **syntax errors are detected during the compilation stage of the code.** Compilation refers to the process of checking and translating the code before it is executed. Since Python is an interpreted language that requires the code to be compiled before it can be executed, syntax errors are detected during compilation and **they are detected before the code is run.** 

Syntax errors are some of the most common errors that developers encounter, and they can often be easily fixed once identified. Python's syntax rules are strict, and the interpreter will flag any code that doesn't follow them with an error message. The error message will typically include a line number and a description of the error, making it easier for the developer to find and fix the issue.

![python.exe](https://drive.google.com/uc?id=1Zo4bPLuiqmEgVSGhtvy9X6h11vkefNcK)

<hr>

Some common syntax errors in Python include:<br> 
- **misspelled or incorrect keywords**,<br> 
- **missing or mismatched parentheses** or<br> 
- **quotes**, and<br> 
- **improperly indented code**<br>

One common mistake is forgetting to close a parentheses or quote, which can cause the interpreter to flag the entire line as an error. Another common mistake is using a keyword as a variable name, which can cause confusion and errors in the code.

<hr>

To avoid syntax errors, it's important to carefully check your code for any mistakes before running it. It's also helpful to use an Integrated Development Environment (IDE) that provides syntax highlighting and error messages in real-time, making it easier to catch mistakes as you write the code.

### <p style="background-color:#0D8D99; font-family:newtimeroman; color:#FFF9ED; font-size:175%; text-align:center; border-radius:10px 10px;">Building a Calculator Using Dict. with Lambda Expression</p>

### We will create a simple calculator using lambda expressions and dictionaries.

We will create a dictionary named **"calculator"** and **its keys will consist of the four arithmetic operator symbols ( +, -, /, *)**

- When we index the `calculator["+"]` key of the dictionary, it will perform addition.
- When we index the `calculator["-"]` key, it will perform subtraction.
- When we index the `calculator["*"]` key, it will perform multiplication.
- When we index the `calculator["/"]` key, it will perform division.

 

for example: `calculator["+"](4, 5)` returns 9

<hr>

### Here's a question that tests your basic Python knowledge:

The value of a dictionary can be a string, an integer, a boolean, a list, a tuple, or even another dictionary. 
Ok. Can the value of a dictionary be a function?

Yes, of course, as long as it returns a value. **If a function returns a value, that value can be used as the value of any key in the dictionary.** Therefore, a function can be used as the value of a key. 

However, **this function should be a single line expression.** In this case, **lambda functions** can be used since they are expressions.<hr>

### Then we will define the key-value pairs like this:

`calculator = {"+" : lambda expression for addition,
              "-" : lambda expression for subtraction,
              "*" : lambda expression for multiplication,
              "/" : lambda expression for division}`

```python
(lambda x, y : x + y) 
```

**I want the "+" key of the dictionary to perform addition when called with (4, 5) and return the result as 9.**

```python
# Calling the lambda function with arguments 4 and 5

(lambda x, y : x + y)(4, 5)
```

```python
dict1 = {"one" : 44,
         "two" : ["ally", "kirby", "betty"]}
```

```python
# We are performing an indexing to access the value of a specific key in a dictionary.

dict1["one"]

# This indexing returned the value associated with the key "one" in the dictionary
```

```python
type(dict1["one"])
```

```python
dict1["two"]

# The indexing returned the value associated with the key 'two' in the dictionary, whatever it may be defined as.
```

```python
type(dict1["two"])
```

```python
dict1 = {"one" : 44,
         "two" : ["ally", "kirby", "betty"],
         "three" : lambda x, y : x+y }
```

<hr>

So, if I define a lambda expression as the value for a key in the dictionary, it will return that lambda expression as the value when the key is accessed.

Please note that at this stage, **since I haven't assigned any arguments to the lambda function, the value won't be an "integer", but rather a "function" that is of the type lambda**

```python
type(lambda x, y : x + y)
```

<hr>

If we index the key of the dictionary in this way, the value returned will be the lambda function itself. 

**If we want the result of the operation performed by the lambda function to be returned, we need to send an argument to the lambda function immediately after indexing, inside parentheses.** 
<br>Like this:

```python
dict1["three"](4,5)
```

**Let's then arrange the dictionary as required :**

```python
calculator = {"+" : (lambda x, y : x + y),
              "-" : (lambda x, y : x - y),
              "*" : (lambda x, y : x * y),
              "/" : (lambda x, y : x / y)}
```

```python
calculator["/"]
```

```python
calculator["/"](9, 3)
```

```python
calculator["-"](14.6, 7.8)
```
