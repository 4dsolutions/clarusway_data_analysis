**Task: Let’s write a program that asks you to enter your exam score and calculates the range in which your degree is based on your exam score.** 
<br>The output would be: e.g, **“Your degree is B+”**
<br><br>95 and above ▶▶ “A+”
<br>90-94 ▶▶ “A”
<br>85-89 ▶▶ “B+”
<br>80-84 ▶▶ “B”
<br>79 and below ▶▶ “below B” or “B-”
<br><br>**Use nested if statement!**

```python
score = int(input("Enter your score :"))

if score >= 90 :
    if score >= 95 :
        score_letter = "A+"
    else :
        score_letter = "A"
        
elif score >= 80 :
    if score >= 85 :
        score_letter = "B+"
    else :
        score_letter = "B"

else :
    score_letter = "B-"
    
print("Your degree is : ", score_letter)
```

## Conditional expressions that return True or False are an essential component of if statements. Therefore, let's recall the operators that return True or False one more time:

**1. logical operators:**

**and, or, not**

**2. Commparision operators:**

**<, >, <=, >=, ==, !=** 

**3. Idendity operators:**

**is ve is not**

**The 'is' operator returns True if two values are the same. The 'is not' operator, on the other hand, returns True if they are not the same, as the opposite of the former statement**


**4. Membership (aitlik) operators.**

**in ve not in**

**The 'in' operator checks whether a value is in a sequence: it returns True if the value is in the sequence, and False otherwise. 'not in' is the opposite: it returns True if the value is not in the sequence, and False if it is.**
