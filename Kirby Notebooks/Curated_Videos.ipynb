In the old days, statistics involved published tables in books and looking up pre-calculated values, such a logs. Confidance intervals could be ascertained based on sample size n, standard deviation, and size of confidence interval required. 

A CI of 95% means: there's only a 5% chance (1 in 20) that the upper-lower range provided by sampling does not include the true value for the whole population. More variance in the sample requires wider CIs.  The bigger the sample, the narrower the CI.
